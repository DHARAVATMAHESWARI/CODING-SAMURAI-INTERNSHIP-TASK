```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Product': ['Apple', 'Banana', 'Carrot', 'Broccoli', 'Milk', 'Cheese'],
    'Category': ['Fruit', 'Fruit', 'Vegetable', 'Vegetable', 'Dairy', 'Dairy'],
    'Units Sold': [50, 80, 40, 30, 60, 20],
    'Price per Unit': [0.5, 0.3, 0.2, 0.4, 1.0, 2.0],
    'Revenue': [25, 24, 8, 12, 60, 40]
}
df = pd.DataFrame(data)

print("Total Revenue:", df['Revenue'].sum())
print("Average Units Sold:", df['Units Sold'].mean())
print("\nRevenue by Category:\n", df.groupby('Category')['Revenue'].sum())

plt.figure(figsize=(8,5))
plt.bar(df['Product'], df['Revenue'], color='skyblue')
plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()

category_revenue = df.groupby('Category')['Revenue'].sum()
plt.figure(figsize=(6,6))
plt.pie(category_revenue, labels=category_revenue.index, autopct='%1.1f%%')
plt.title('Revenue Share by Category')
plt.show()
```
